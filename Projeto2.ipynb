## Introdução 


### Contexto do tema "Qualidade do Vinho"

> O vinho é uma bebida mundialmente apreciada. O aroma anuncia o sabor do vinho, que, quando em contato com as papilas gustativas, ativam nosso paladar e aguça-o. Antes de experimentá-lo, porém, devemos estar prontos para distinguir os variados tipos e saber com qual comida cada um combina cada um, para ter harmonia entre os sabores.


#### O que é um vinho?
O vinho é uma bebida natural. Por definição, ele é resultado da fermentação das uvas, que acontece quando as leveduras (fungos microscópicos), presentes nas cascas das uvas, penetram no fruto e transformam os açúcares da uva (glicose e frutose) em álcool etílico e gás carbônico:
> Açúcar da uva + leveduras = Álcool etílico + Gás Carbônico (CO2) + Energia

#### Como classificar um vinho? 
Um vinho pode ser classificado de acordo com a classe, cor, teor de açúcar e tipos.

- **Classe**

>> - Vinhos de mesa: variam entre finos, nobres, especiais, comuns, frisantes ou gaseificados, com graduação alcoólica de 10 a 13°

>> - Vinhos leves: graduação alcoólica de 7 a 9,9°

>> - Champagne: é um tipo espumante e com gradação alcoólica de 14 a 18°

>> - Vinhos compostos: são preparados com adição de algum outro ingrediente, como produtos de origem animal, vegetal ou óleos, com graduação alcoólica de 15 a 18°

>> - Vinhos licorosos: como o nome sugere, tem aspecto de licor, com graduação alcoólica de 15 a 18°

- **Cor**

>> **Vinho tinto:** é o tipo de vinho mais comercializado no mundo. É elaborado a partir de variedades de uvas de coloração avermelhada ou roxa.

>> **Vinho branco:** elaborado a partir da fermentação de uvas brancas ou escuras, sem uso da casca.

>> **Vinho rosé:** pode ser uma combinação da preparação tinta com a preparação branca, mas também pode ser elaborado com uvas escuras que liberem pouca coloração no processo de fermentação.

- **Teor de açúcar**

>> **Vinho seco (brut):** a cada litro é adicionado até 5g de açúcar no preparo.

>> **Vinho demi-sec:** a cada litro é adicionado entre 5 e 20g de açúcar no preparo.

>> **Vinho suave:** a cada litro é adicionado mais de 20g de açúcar no preparo.

- **Tipos de uvas (e de vinhos!)**

>> **Cabernet Sauvignon:** é o tipo de uva mais utilizado na produção de vinhos no mundo todo. Os vinhos produzidos são, geralmente, de mesa, tintos e secos e equilibrados no amargor e no aroma. Harmoniza bem com queijos fortes e carnes vermelhas.

>> **Malbec:** tipo de vinho mais encorpado e amargo que o Cabernet. Par perfeito para o churrasco e um bom pedaço de picanha.

>> **Merlot:** é semelhante à Cabernet, porém proporciona um aroma mais frutado e sofisticado. Combina bem com massas, molhos encorpados e aves.

>> **Tannat:** possui coloração intensa e sabor característico. Esse vinho dá match com gorduras e pratos bastante encorpados e condimentados.

>> **Pinot Noir:** pode ser usada para elaborar vinhos de mesa ou espumantes, pois é suave e muito delicada. Adequado para queijos leves, carne de porco, salmão e frutos do mar.

>> **Chardonnay:** tipo de uva branca, considerada uma das mais nobres na elaboração de vinhos de mesa e espumantes. Também é usada na elaboração de champagnes. Pratos à base de aves, como galinha ou peru; e peixes ou frutos do mar formam uma ótima pedida para um vinho Chardonnay.

## Exploração dos Dados:

> Nessa primeira interação com os dados, optamos por realizar a organização dos dados, entender o que é relevante para está no nosso dataset. 

>A organização das colunas é fundamental para uma boa visualização e interpretação dos nossos dados, realizando também a exclusão da coluna Id, que é irrelevante para nosso projeto. 


> Nesse sentido, iremos realizar os cálculos das covariâncias, plotando os gráficos catterplots e histogramas das colunas individualmente para que possamos analisar os dados com mais cuidado e entendermos a relação que os features tem diretamente com a nossa target, antes de realizarmos essa comparação em conjunto.

1) **Organizando as colunas que serão utilizadas:**

```python
dados_fia = dados['fixed acidity']
dados_voa = dados['volatile acidity']
dados_cta = dados['citric acid']
dados_res = dados['residual sugar']
dados_chl = dados['chlorides']
dados_fsd = dados['free sulfur dioxide']
dados_tsd = dados['total sulfur dioxide']
dados_den = dados['density']
dados_pH = dados['pH']
dados_sul = dados['sulphates']
dados_alc = dados['alcohol']
dados_qua = dados['quality']
```

2) **Calculando as covariâncias e verificando seus valores para poder verificar a correlação entre as colunas e a variável Target "quality":**

```python
cov_fia_qua = dados_fia.cov(dados_qua)
cov_voa_qua = dados_voa.cov(dados_qua)
cov_cta_qua = dados_cta.cov(dados_qua)
cov_res_qua = dados_res.cov(dados_qua)
cov_chl_qua = dados_chl.cov(dados_qua)
cov_fsd_qua = dados_fsd.cov(dados_qua)
cov_tsd_qua = dados_tsd.cov(dados_qua)
cov_den_qua = dados_den.cov(dados_qua)
cov_pH_qua = dados_pH.cov(dados_qua)
cov_sul_qua = dados_sul.cov(dados_qua)
cov_alc_qua = dados_alc.cov(dados_qua)

print(cov_fia_qua)
print(cov_voa_qua)
print(cov_cta_qua)
print(cov_res_qua)
print(cov_chl_qua)
print(cov_fsd_qua)
print(cov_tsd_qua)
print(cov_den_qua)
print(cov_pH_qua)
print(cov_sul_qua)
print(cov_alc_qua)
```

3) **Desvios padrões e Correlações:**

```python
DP_dados_fia = dados_fia.std() 
DP_dados_voa = dados_voa.std()
DP_dados_cta = dados_cta.std()
DP_dados_res = dados_res.std()
DP_dados_chl = dados_chl.std()
DP_dados_fsd = dados_fsd.std()
DP_dados_tsd = dados_tsd.std()
DP_dados_den = dados_den.std()
DP_dados_pH = dados_pH.std()
DP_dados_sul = dados_sul.std()
DP_dados_alc = dados_alc.std()
DP_dados_qua = dados_qua.std()

Cor_fia_qua = cov_fia_qua/(DP_dados_fia*DP_dados_qua) 
Cor_voa_qua = cov_voa_qua/(DP_dados_voa*DP_dados_qua)
Cor_cta_qua = cov_cta_qua/(DP_dados_cta*DP_dados_qua)
Cor_res_qua = cov_res_qua/(DP_dados_res*DP_dados_qua)
Cor_chl_qua = cov_chl_qua/(DP_dados_chl*DP_dados_qua) 
Cor_fsd_qua = cov_fsd_qua/(DP_dados_fsd*DP_dados_qua)
Cor_tsd_qua = cov_tsd_qua/(DP_dados_tsd*DP_dados_qua)
Cor_den_qua = cov_den_qua/(DP_dados_den*DP_dados_qua)
Cor_pH_qua = cov_pH_qua/(DP_dados_pH*DP_dados_qua)
Cor_sul_qua = cov_sul_qua/(DP_dados_sul*DP_dados_qua)
Cor_alc_qua = cov_alc_qua/(DP_dados_alc*DP_dados_qua)

# Imprimindo as correlações
print(Cor_fia_qua)
print(Cor_voa_qua)
print(Cor_cta_qua)
print(Cor_res_qua)
print(Cor_chl_qua)
print(Cor_fsd_qua)
print(Cor_tsd_qua)
print(Cor_den_qua)
print(Cor_pH_qua)
print(Cor_sul_qua)
print(Cor_alc_qua)
```

Através das correlações exibidas acima, é possível perceber que a acidez volátil, a quantidade de cloretos, dióxido de enxofre livre, total de dióxido de enxofre, densidade e pH são inversamente proporcionais à qualidade do vinho. Além disso, os cloretos são muito prejudiciais para a qualidade do vinho, como pode ser percebida pela correlação de -0.40, que é bem maior do que os outros dados.

Sobre as correlações positivas, a quantidade de álcool é diretamente proporcional à qualidade do vinho, sendo seu valor 0.48.

## Método 'Pearson':
>O coeficiente de correlação de Pearson é um teste que mede a relação estatística entre duas variáveis contínuas. Se a associação entre os elementos não for linear, o coeficiente não será representado adequadamente.


>O coeficiente de correlação de Pearson pode ter um intervalo de valores de +1 a -1. Um valor de 0 indica que não há associação entre as duas variáveis. Um valor maior que 0 indica uma associação positiva. Isto é, à medida que o valor de uma variável aumenta, o mesmo acontece com o valor da outra variável. Um valor menor que 0 indica uma associação negativa. Isto é, à medida que o valor de uma variável aumenta, o valor da outra diminui.

```python
# Fazendo a correlação de todas as colunas com o método Pearson:
novo.corr(method ='pearson')
```

#### Análise do uso do método:
> Dado que nosso interesse está em análisar a correlação da **Qualidade do vinho com as demais colunas**, a descrição do uso do método será feito de acordo com a correlação da **Linha de Qualidade** e todas as colunas.
>> O que podemos dizer com estes dados apresentados é que a colunas: **acidez, acidezCitrica, AçucarResidual, sulfatos e álcool** nos mostram uma associação positiva em relação a **Qualidade do vinho**. Por outro lado as colunas: **volatilidadeAcida, cloretos, DioxDeEnxofreLivre, dioxDeEnxofreTotal, densidade, pH** nos mostram uma associação negativa em relação a **Qualidade do vinho**.

> Com essa análise é notavél ver dizer que: a acidez, acidezCitrica, AçucarResidual, sulfatos e álcool tem uma influência positiva quando o quesito é a qualidade dos vinhos. Já a volatilidadeAcida, cloretos, DioxDeEnxofreLivre, dioxDeEnxofreTotal, densidade, pH, tem uma influêcia negativa, dado a qualidade.

<h2> Primeiro Modelo - Predição por Árvore de Decisão </h2>

> Como descrito na documentação da biblioteca "Scikit-Learn", as Árvores de Decisão são um método de aprendizado supervisionado não paramétrico usado para classificação. O objetivo é criar um modelo que preveja o valor de uma variável de destino aprendendo regras de decisão simples inferidas dos recursos de dados. Uma árvore pode ser vista como uma aproximação constante por partes.

<img src="arvore_decisão"></img>

1) Primeiramente, separamos o nosso dataframe entre features e target. Fizemos isso, pois a função "train_test_split" realiza a separação da nossa tabela em duas, onde uma é a de treino e a outra é a de teste. Essas tabelas serão ultilizadas para "ensinar" o classificador e testá-lo posteriormente.

```python
vinho_vars = novo.drop(['Qualidade'], axis=1)
```

2) Separando o nosso dataframe em 2, um para treinar o nosso classificadore outro para testar ele.

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(vinho_vars, novo["Qualidade"], test_size = 0.4, random_state = 42)
```

3) Após isso, utilizamos as funçoes "DecisionTreeClassifier().fit()" do "Sklearn" para criar nossa árvore de decisões. A função "fit" recebe como argumentos uma tabela de features e seus respectivos targets para treinar uma árvore de decisões criada pela função "DecisionTreeClassifier". Logo após, utilizamos a árvore treinada para prever qual o target dos vinhos presentes em uma tabela de teste pelos valores presentes em suas features. Concluído isso, comparamos os resultados reais com os de previsão.

```python
decision_maker = DecisionTreeClassifier().fit(X_treinamento,y_treinamento)

y_previsto = decision_maker.predict(X_teste)

y_previsto = pd.Series(y_previsto)

print(classification_report(y_teste, y_previsto))
```

4) Utilizamos então a mesma árvore de decisão para prever quais deveriam ser os targets dos vinhos da planilha que o código usou para treino, para entender o quanto ele aprendeu da planilha de treinamento. Assim, comparamos o real com o previsto

```python
y_prev_treino = decision_maker.predict(X_treinamento)
pd.crosstab(y_treinamento, y_prev_treino)
```

```python
print(classification_report(y_treinamento, y_prev_treino))
```

5) Para entender mais qual a eficiência de nosso classificador, calculamos o RMSE de ambas as previsões. o RMSE (Root Mean Squared Error) é uma medida de erro absoluto que nos permite entender como nossa árvore de decisões está errando.

```python
rmse_teste = math.sqrt(mean_squared_error(y_teste, y_previsto))

rmse_teste
```

```python
rmse_treino = math.sqrt(mean_squared_error(y_treinamento, y_prev_treino))

rmse_treino
```

6) Por fim, realizamos uma tabela que indica as relevâncias de cada uma das features para compreendermos o quanto cada uma dessas variáveis afetam na classificação da qualidade.

```python
relevancia_features = pd.Series(data = decision_maker.feature_importances_, index = vinho_vars.columns)
plt.figure(figsize = (10, 8))
sns.barplot(x = relevancia_features, y = relevancia_features.index, orient='h').set_title('Importância de cada feature para a qualidade do vinho')
plt.show()
```

<h3> Estatísticas da Árvore de Decisão: </h3>

7) Por fim, utilizamos a função "score" tanto para calcular acurácia da previsão de treino como para a previsão de teste. Essa função calcula a taxa de acerto do nosso classificador. Além disso, fizemos o processo de Validação Cruzada utilizando 10 ciclos e calculamos a performance média e desvio padrão das duas medidas R2.

```python
score_de_treino = decision_maker.score(X_treinamento, y_treinamento)
score_de_test = decision_maker.score(X_teste, y_teste)
f'A acurácia do classificador da árvore de decisão, ao analisar a base de treinamento, é de {score_de_treino:.4f} e ao analisar a base de teste é de {score_de_test:.4f}, assim como já tinha sido concluido anteriormente'
```

>R² é o percentual de variação na resposta que é prevista pelo modelo, quanto mais próximo de 1, melhor, pois significa que o modelo se adequa bem a diferentes conjuntos de dados.

```python
kfold  = KFold(n_splits=10, shuffle=True)
valores_R2 = cross_val_score(decision_maker, vinho_vars, dados['quality'], cv = kfold)

for i in valores_R2:
    print("Score do R2 : {0}".format(i))
print("Média dos R2: {0}".format(valores_R2.mean()))
print("Desvio padrão dos R2: {0}".format(valores_R2.std()))
```

>A técnica da árvore de decisão apresenta um desvio padrão alto, próximo de 4.9%, o que indica uma variação que pioraria consideravelmente sua performance, em torno da média de aproximadamente 57% do R².

<h3> Análise do classificador</h3>

>- O nosso classificador por árvore de decisão apresentou a princípio uma taxa de acerto de 56%, um valor bastante razoável considerando a quantidade de targets e features da nossa base de dados. Porém, podemos ir um pouco mais a fundo e concluir algumas coisas sobre essa taxa de erro. 

>- É fácil concluir que o nosso modelo de predição não é eficaz para determinar vinhos no qual a qualidade seja muito alta ou muito baixa. Isso ocorre, pois o dataframe base do projeto não possuia muitas amostras de vinhos de qualidade muito baixa, como 3 e 4, e também apresentava poucas informações de vinhos de qualidade muito alta, como qualidade 8. Como consequência disso, a taxa de acerto do nosso classificador para qualidades "extremas" cai muito quando comparado à taxa de acerto para qualidades intermediárias.

>- Além disso, quando se faz a análise do valor de RMSE (Raiz quadrada do erro-médio), observa-se que o nosso classificador, apesar de errar, costumava errar com valores próximos. Por exemplo, quando o nosso classificador ia definir a qualidade de um vinho que se enquadra como 5, ele acabava por atribuílo a um vinho de qualidade 6, que é próximo, mas é equivocado. Pode-se concluir isso já que o valor do RMSE é aproximadamente 0.84, ou seja, o classificador costuma errar o valor em 0.84. Isso é confirmado quando analisamos o crosstab entre as qualidades reais de cada vinho e as qualidades atribuidas pelo codigo da arvore de decisão.

>- Por fim, observamos que as 3 colunas mais relevantes na decisão da qualidade do vinho foram em ordem de prioridade: Quantidade de álcool no vinho, Dióxido de Enxofre Total e Quantidade de sulfatos.

>- Em um último momento, percebemos que o nosso classificador provavelmente se enquadraria na classificação de overfitting. Pode-se afirmar isso uma vez que o classificador aprendeu perfeitamente o padrão dos dados analisados, não cometendo nenhum erro quando necessário fazer a predição dos vinhos da coluna de treinamento, como conclui-se com o valor do RMSE igual a 0, porém, ao que parece, ele não consegue se adaptar muito bem a novas condições, o que acaba ocasionando em erros de previsão, classificando um vinho com uma qualidade incorreta quase 40% das vezes.

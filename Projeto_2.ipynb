## Introdução 


### Contexto do tema "Qualidade do Vinho"

> O vinho é uma bebida mundialmente apreciada. O aroma anuncia o sabor do vinho, que, quando em contato com as papilas gustativas, ativam nosso paladar e aguça-o. Antes de experimentá-lo, porém, devemos estar prontos para distinguir os variados tipos e saber com qual comida cada um combina cada um, para ter harmonia entre os sabores.


#### O que é um vinho?
O vinho é uma bebida natural. Por definição, ele é resultado da fermentação das uvas, que acontece quando as leveduras (fungos microscópicos), presentes nas cascas das uvas, penetram no fruto e transformam os açúcares da uva (glicose e frutose) em álcool etílico e gás carbônico:
> Açúcar da uva + leveduras = Álcool etílico + Gás Carbônico (CO2) + Energia

#### Como classificar um vinho? 
Um vinho pode ser classificado de acordo com a classe, cor, teor de açúcar e tipos.

- **Classe**

>> - Vinhos de mesa: variam entre finos, nobres, especiais, comuns, frisantes ou gaseificados, com graduação alcoólica de 10 a 13°

>> - Vinhos leves: graduação alcoólica de 7 a 9,9°

>> - Champagne: é um tipo espumante e com gradação alcoólica de 14 a 18°

>> - Vinhos compostos: são preparados com adição de algum outro ingrediente, como produtos de origem animal, vegetal ou óleos, com graduação alcoólica de 15 a 18°

>> - Vinhos licorosos: como o nome sugere, tem aspecto de licor, com graduação alcoólica de 15 a 18°

- **Cor**

>> **Vinho tinto:** é o tipo de vinho mais comercializado no mundo. É elaborado a partir de variedades de uvas de coloração avermelhada ou roxa.

>> **Vinho branco:** elaborado a partir da fermentação de uvas brancas ou escuras, sem uso da casca.

>> **Vinho rosé:** pode ser uma combinação da preparação tinta com a preparação branca, mas também pode ser elaborado com uvas escuras que liberem pouca coloração no processo de fermentação.

- **Teor de açúcar**

>> **Vinho seco (brut):** a cada litro é adicionado até 5g de açúcar no preparo.

>> **Vinho demi-sec:** a cada litro é adicionado entre 5 e 20g de açúcar no preparo.

>> **Vinho suave:** a cada litro é adicionado mais de 20g de açúcar no preparo.

- **Tipos de uvas (e de vinhos!)**

>> **Cabernet Sauvignon:** é o tipo de uva mais utilizado na produção de vinhos no mundo todo. Os vinhos produzidos são, geralmente, de mesa, tintos e secos e equilibrados no amargor e no aroma. Harmoniza bem com queijos fortes e carnes vermelhas.

>> **Malbec:** tipo de vinho mais encorpado e amargo que o Cabernet. Par perfeito para o churrasco e um bom pedaço de picanha.

>> **Merlot:** é semelhante à Cabernet, porém proporciona um aroma mais frutado e sofisticado. Combina bem com massas, molhos encorpados e aves.

>> **Tannat:** possui coloração intensa e sabor característico. Esse vinho dá match com gorduras e pratos bastante encorpados e condimentados.

>> **Pinot Noir:** pode ser usada para elaborar vinhos de mesa ou espumantes, pois é suave e muito delicada. Adequado para queijos leves, carne de porco, salmão e frutos do mar.

>> **Chardonnay:** tipo de uva branca, considerada uma das mais nobres na elaboração de vinhos de mesa e espumantes. Também é usada na elaboração de champagnes. Pratos à base de aves, como galinha ou peru; e peixes ou frutos do mar formam uma ótima pedida para um vinho Chardonnay.

Organizando as colunas que serão utilizadas

```python
dados_fia = dadoss['fixed acidity']
dados_voa = dadoss['volatile acidity']
dados_cta = dadoss['citric acid']
dados_res = dadoss['residual sugar']
dados_chl = dadoss['chlorides']
dados_fsd = dadoss['free sulfur dioxide']
dados_tsd = dadoss['total sulfur dioxide']
dados_den = dadoss['density']
dados_pH = dadoss['pH']
dados_sul = dadoss['sulphates']
dados_alc = dadoss['alcohol']
dados_qua = dadoss['quality']
```

Calculando as covariâncias e verificando seus valores para poder verificar a correlação entre as colunas e a variável Target "quality"

```python
cov_fia_qua = dados_fia.cov(dados_qua)
cov_voa_qua = dados_voa.cov(dados_qua)
cov_cta_qua = dados_cta.cov(dados_qua)
cov_res_qua = dados_res.cov(dados_qua)
cov_chl_qua = dados_chl.cov(dados_qua)
cov_fsd_qua = dados_fsd.cov(dados_qua)
cov_tsd_qua = dados_tsd.cov(dados_qua)
cov_den_qua = dados_den.cov(dados_qua)
cov_pH_qua = dados_pH.cov(dados_qua)
cov_sul_qua = dados_sul.cov(dados_qua)
cov_alc_qua = dados_alc.cov(dados_qua)

print(cov_fia_qua)
print(cov_voa_qua)
print(cov_cta_qua)
print(cov_res_qua)
print(cov_chl_qua)
print(cov_fsd_qua)
print(cov_tsd_qua)
print(cov_den_qua)
print(cov_pH_qua)
print(cov_sul_qua)
print(cov_alc_qua)
```

Desvios padrões e Correlações

```python
DP_dados_fia = dados_fia.std() 
DP_dados_voa = dados_voa.std()
DP_dados_cta = dados_cta.std()
DP_dados_res = dados_res.std()
DP_dados_chl = dados_chl.std()
DP_dados_fsd = dados_fsd.std()
DP_dados_tsd = dados_tsd.std()
DP_dados_den = dados_den.std()
DP_dados_pH = dados_pH.std()
DP_dados_sul = dados_sul.std()
DP_dados_alc = dados_alc.std()
DP_dados_qua = dados_qua.std()

Cor_fia_qua = cov_fia_qua/(DP_dados_fia*DP_dados_qua) 
Cor_voa_qua = cov_voa_qua/(DP_dados_voa*DP_dados_qua)
Cor_cta_qua = cov_cta_qua/(DP_dados_cta*DP_dados_qua)
Cor_res_qua = cov_res_qua/(DP_dados_res*DP_dados_qua)
Cor_chl_qua = cov_chl_qua/(DP_dados_chl*DP_dados_qua) 
Cor_fsd_qua = cov_fsd_qua/(DP_dados_fsd*DP_dados_qua)
Cor_tsd_qua = cov_tsd_qua/(DP_dados_tsd*DP_dados_qua)
Cor_den_qua = cov_den_qua/(DP_dados_den*DP_dados_qua)
Cor_pH_qua = cov_pH_qua/(DP_dados_pH*DP_dados_qua)
Cor_sul_qua = cov_sul_qua/(DP_dados_sul*DP_dados_qua)
Cor_alc_qua = cov_alc_qua/(DP_dados_alc*DP_dados_qua)

# Imprimindo as correlações
print(Cor_fia_qua)
print(Cor_voa_qua)
print(Cor_cta_qua)
print(Cor_res_qua)
print(Cor_chl_qua)
print(Cor_fsd_qua)
print(Cor_tsd_qua)
print(Cor_den_qua)
print(Cor_pH_qua)
print(Cor_sul_qua)
print(Cor_alc_qua)
```

Através das correlações exibidas acima, é possível perceber que a acidez volátil, a quantidade de cloretos, dióxido de enxofre livre, total de dióxido de enxofre, densidade e pH são inversamente proporcionais à qualidade do vinho. Além disso, os cloretos são muito prejudiciais para a qualidade do vinho, como pode ser percebida pela correlação de -0.40, que é bem maior do que os outros dados.

Sobre as correlações positivas, a quantidade de álcool é diretamente proporcional à qualidade do vinho, sendo seu valor 0.48.

## Método 'Pearson':
>O coeficiente de correlação de Pearson é um teste que mede a relação estatística entre duas variáveis contínuas. Se a associação entre os elementos não for linear, o coeficiente não será representado adequadamente.


>O coeficiente de correlação de Pearson pode ter um intervalo de valores de +1 a -1. Um valor de 0 indica que não há associação entre as duas variáveis. Um valor maior que 0 indica uma associação positiva. Isto é, à medida que o valor de uma variável aumenta, o mesmo acontece com o valor da outra variável. Um valor menor que 0 indica uma associação negativa. Isto é, à medida que o valor de uma variável aumenta, o valor da outra diminui.

```python
# Fazendo a correlação de todas as colunas com o método Pearson:
novo.corr(method ='pearson')
```

#### Análise do uso do método:
> Dado que nosso enteresse está em análisar a correlação da **Qualidade do vinho com as demais colunas**, a descrição do uso do método será feito de acordo com a correlação da **Linha de Qualidade** e todas as colunas.
>> O que podemos dizer com estes dados apresentados é que a colunas: **acidez, acidezCitrica, AçucarResidual, sulfatos e álcool** nos mostram uma associação positiva em relação a **Qualidade do vinho**. Por outro lado as colunas: **volatilidadeAcida, cloretos, DioxDeEnxofreLivre, dioxDeEnxofreTotal, densidade, pH** nos mostram uma associação negativa em relação a **Qualidade do vinho**.

> Com essa análise é notavél ver dizer que: a acidez, acidezCitrica, AçucarResidual, sulfatos e álcool tem uma influência positiva quando o quesito é a qualidade dos vinhos. <h2>Agora vamos começar a fazer um método de predição chamado Regressão Logística

Primeiramente removemos a coluna qualidade do DataFrame pois ela é a variável Target.
Depois colocamos as colunas restantes em um novo DataFrame oeganizado pelo método StandartScaler.
Por fim separamos a coluna qualidade em uma outra variável chamada resultados_p para uso posterior.

```python
dados_explicativos = novo.drop("Qualidade",axis=1)
dados_explicativos_ajustados = StandardScaler().fit_transform(dados_explicativos)
resultados_p = LabelEncoder().fit_transform(novo.Qualidade)
novo.Qualidade
```

Após isso, criamos as variáveis X_train, X_test, y_train e y_test utilizando a função train_test_split.

```python
X_train, X_test, y_train, y_test = train_test_split(dados_explicativos_ajustados, resultados_p, test_size = 0.3, random_state=42)
```

Escolhemos o método de Regressão Logística abaixo para efetuar a predição:

```python
classificador_logistico = LogisticRegression()
```

```python
classificador_logistico.fit(X_train,y_train)
```

```python
ypred = classificador_logistico.predict(X_test)
ypred
```

```python
print(classification_report(y_test, ypred))
```

```python
coefs = np.abs(classificador_logistico.coef_[0])
indices = np.argsort(coefs)[::-1]
plt.figure()
plt.title("Importância de cada feature - Regressão Logística")
plt.bar(range(10), coefs[indices[:10]],
       color="violet", align="center")
plt.xticks(range(10), dados_explicativos.columns[indices[:10]], rotation=45, ha='right')
plt.subplots_adjust(bottom=0.3)
```

```python
rmse_teste = math.sqrt(mean_squared_error(y_test, ypred))

rmse_teste
```

```python
y_prev_treino = classificador_logistico.predict(X_train)
rmse_treino = math.sqrt(mean_squared_error(y_train, y_prev_treino))

rmse_treino
```

```python
kfold  = KFold(n_splits=10, shuffle=True)
valores_R2 = cross_val_score(classificador_logistico, dados_explicativos, vinho['quality'], cv = kfold)

for i in valores_R2:
    print("Score do R2 : {0}".format(i))
print("Média dos R2: {0}".format(valores_R2.mean()))
print("Desvio padrão dos R2: {0}".format(valores_R2.std()))
```
